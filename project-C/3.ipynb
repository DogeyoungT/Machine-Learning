```python
import pandas as pd
import numpy as np
user_info_df = pd.read_csv('data_movie_lens_100k/user_info.csv')
user_info = (
    user_info_df['user_id'].values,
    user_info_df['age'].values,
    user_info_df['is_male'].values)
```

```python
x_tr = model_50.param_dict['U']
y_tr = user_info[2]
```

```python
from sklearn.neural_network import MLPClassifier
mpl = MLPClassifier(
        hidden_layer_sizes=50,
        activation='relu',
        alpha=0.0001,
        tol=1e-5,
        n_iter_no_change=50,
        max_iter=2000,
        solver='sgd',
        batch_size=100,
        learning_rate='adaptive', learning_rate_init=0.3, momentum=0.0,
        )
mpl.fit(x_tr, y_tr)
```

```python
from sklearn.metrics import confusion_matrix
y_h = mpl.predict(x_tr)
confusion_matrix(y_tr, y_h)
```

```python
import sklearn
mpl = MLPClassifier(
        hidden_layer_sizes=50,
        activation='relu',
        alpha=0.0001,
        tol=1e-5,
        n_iter_no_change=50,
        max_iter=2000,
        solver='sgd',
        batch_size=100,
        learning_rate='adaptive', learning_rate_init=0.3, momentum=0.0,
        )
mpl_hyperparameter_grid_by_name = dict(
    hidden_layer_sizes = [1,10,50],
    learning_rate_init = [0.1,0.3,0.5]
    )
mpl_searcher = sklearn.model_selection.GridSearchCV(
    mpl,
    mpl_hyperparameter_grid_by_name,
    scoring = 'balanced_accuracy',
    return_train_score=True,
    refit=False)
mpl_searcher.fit(x_tr, y_tr)
mpl_search_results_df = pd.DataFrame(mpl_searcher.cv_results_).copy()
```

```python
print(mpl_searcher.best_params_)
print(mpl_searcher.best_score_)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import balanced_accuracy_score
```

```python
kf = KFold(n_splits=3)
for train_index, test_index in kf.split(x_tr):

    X_train, X_test = x_tr[train_index], x_tr[test_index]
    y_train, y_test = y_tr[train_index], y_tr[test_index]
    mpl = MLPClassifier(
        hidden_layer_sizes=10,
        activation='relu',
        alpha=0.001,
        tol=1e-5,
        n_iter_no_change=50,
        max_iter=3000,
        solver='sgd',
        batch_size=100,
        learning_rate='adaptive', learning_rate_init=0.1, momentum=0.0,
        )
    mpl.fit(X_train, y_train)
    y_h = mpl.predict(X_test)
    print(confusion_matrix(y_h, y_test))
    print(balanced_accuracy_score(y_h, y_test))
```
